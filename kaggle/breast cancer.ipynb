```python
import pandas as pd
df_train = pd.read_csv('../data/breast-cancer-train.csv')
df_test = pd.read_csv('../data/breast-cancer-test.csv')

#print(len(df_train))

df_test_negative = df_test.loc[df_test['Type'] == 0][['Clump Thickness','Cell Size']]
df_test_positive = df_test.loc[df_test['Type'] == 1][['Clump Thickness','Cell Size']]

#print(df_test_negative['Clump Thickness'][:10])

import matplotlib.pyplot as plt
plt.scatter(df_test_negative['Clump Thickness'],df_test_negative['Cell Size'],marker='o',s=200,c='red')
plt.scatter(df_test_positive['Clump Thickness'],df_test_positive['Cell Size'],marker='x',s=150,c='black')
plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')
plt.show()
```

```python
import numpy as np
intercept = np.random.random(1) # 截距
coef = np.random.random(2) # 系数
#print(intercept)
#print(coef)

lx = np.arange(0,12)
ly = (-intercept - lx * coef[0]) / coef[1]  # ax + by + c = z 当z=0时，变成一条直线  y = (-c-ax)/b
#print(lx)
plt.plot(lx,ly,c='yellow')
plt.scatter(df_test_negative['Clump Thickness'],df_test_negative['Cell Size'],marker='o',s=200,c='red')
plt.scatter(df_test_positive['Clump Thickness'],df_test_positive['Cell Size'],marker='x',s=150,c='black')
plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(df_train[['Clump Thickness','Cell Size']][:10], df_train['Type'][:10])
print('Tesing accuracy(10 training samples):',lr.score(df_test[['Clump Thickness','Cell Size']],df_test['Type']))
```

```python
intercept = lr.intercept_
print(lr.coef_)
coef = lr.coef_[0,:]  # 取二维数组 所有列第0个数据  [:,0] 所有行的第0个数据
print(coef)
lx = np.arange(0,12)
ly = (-intercept - lx * coef[0]) / coef[1]  # ax + by + c = z 当z=0时，变成一条直线  y = (-c-ax)/b
#print(lx)
plt.plot(lx,ly,c='yellow')
plt.scatter(df_test_negative['Clump Thickness'],df_test_negative['Cell Size'],marker='o',s=200,c='red')
plt.scatter(df_test_positive['Clump Thickness'],df_test_positive['Cell Size'],marker='x',s=150,c='black')
plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')
plt.show()
```
